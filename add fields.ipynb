```python
import json

def process_and_save_json(input_path, output_path, second_path=None):
    # Read the JSON data from the input file
    with open(input_path, 'r') as file:
        data = json.load(file)

    # Loop through the data and process as required
    for user_key, user_data in data.items():
        user_data['annotator_id'] = None
        for key, value in user_data.items():
            # Check if the key is a digit
            if key.isdigit():
                # Add the 'requires_rewrite' key
                user_data[key]['requires rewrite'] = None

                # Loop through each rewrite object
                for rewrite_type in ['g_rewrite', 'p_rewrite', 'q_rewrite']:
                    rewrite_data = value.get(rewrite_type)
                    if rewrite_data:
                        # Add 'optimal' key next to 'score'
                        rewrite_data['optimal'] = None

    # Write the modified data to the output file
    with open(output_path, 'w') as file:
        json.dump(data, file, indent=4)

    if second_path is not None:
        try:
            with open(second_path, 'w') as file:
                json.dump(data, file, indent=4)
        except:
            print('folder doesnt exist')

# Define the input and output paths
input_path = 'annotation_dict.json'
output_path = 'target.json'
second_output_path = 'dist\\target.json'

# Process the JSON data and save it
process_and_save_json(input_path, output_path, second_path=second_output_path)
```

```python
import json
import os

def find_nulls(input_path):
    # Print the name of the file being processed
    print(f'Processing file: {input_path}')

    # Read the JSON data from the input file
    with open(input_path, 'r') as file:
        data = json.load(file)

    def check_for_null_value(data, parent_keys=[]):
        # Check if the data is a dictionary and iterate through its items
        if isinstance(data, dict):
            for key, value in data.items():
                # Recursive call for nested dictionaries with updated parent_keys
                check_for_null_value(value, parent_keys + [key])
        else:
            # Check for null value (None in Python) directly in the data
            if data is None:
                print(f'Null found in {input_path} at {" -> ".join(parent_keys)}')

    # Loop through the data and check for null values
    for dialog_key, dialog_data in data.items():
        for turn_key, turn_data in dialog_data.items():
            if turn_key == '1':
                continue
            check_for_null_value(turn_data, [dialog_key, turn_key])

def check_for_null(folder_path):
    # Iterate over each file in the specified folder
    for file_name in os.listdir(folder_path):
        # Check if the file is a JSON file
        if file_name.endswith('.json'):
            file_path = os.path.join(folder_path, file_name)
            # Call find_nulls for each JSON file
            find_nulls(file_path)

# Define the folder path containing JSON files (update this with the actual path)
folder_path = f'C:\\OneAI\\OneDrive - OneAI\\annotation\\rewrite\\dist\\rewrite_softwares\\annotations'

# Call the function with the correct argument
check_for_null(folder_path)
```

```python
import json
import os
import pandas as pd

def find_requires_rewrite_zero(input_path):
    # Print the name of the file being processed
    print(f'Processing file: {input_path}')

    # Read the JSON data from the input file
    with open(input_path, 'r') as file:
        data = json.load(file)

    def print_original_question(dialog_data, turn_num):
        for dialog_turn_data in dialog_data:
            if dialog_turn_data['turn_num'] != turn_num:
                continue

            print(f"Original Question: {dialog_turn_data['original_question']}\n")
        
    def print_rewrites(turn_data):
        for rewrite_key, rewrite_data in turn_data.items():
            if isinstance(rewrite_data, dict) and rewrite_data['optimal'] == 1:
                print(f"\t[{rewrite_key}]: {rewrite_data['text']}\t[{rewrite_data['score']}][{rewrite_data['optimal']}]")

    def optimal_exists(turn_data):
        for rewrite_key, rewrite_data in turn_data.items():
            if isinstance(rewrite_data, dict) and rewrite_data['optimal'] == 1:
                return True
            return False

    def print_annotator_rewrite(turn_data):
        if turn_data['annotator rewrite'] == -1:
            print('\nAnnotator Rewrite: **no rewrite**')

        else:
            print(f"\nAnnotator Rewrite: {turn_data['annotator rewrite']}")

    # Loop through the data and check for null values
    for dialog_key, dialog_data in data.items():
        for turn_key, turn_data in dialog_data.items():
            if turn_key == '1':
                continue
            if turn_key.isdigit():
                if turn_data['requires rewrite'] == 0 and optimal_exists(turn_data):
                    print('--------------------')
                    print_original_question(dialog_data['dialog'], int(turn_key))
                    print_rewrites(turn_data)
                    print_annotator_rewrite(turn_data)

def check_for_zero(folder_path):
        # Iterate over each file in the specified folder
        for file_name in os.listdir(folder_path):
            # Check if the file is a JSON file
            if file_name.endswith('.json'):
                file_path = os.path.join(folder_path, file_name)
                # Call find_nulls for each JSON file
                find_requires_rewrite_zero(file_path)



# Define the folder path containing JSON files (update this with the actual path)
folder_path = r"C:\OneAI\OneDrive - OneAI\annotation\rewrite\AnnotatorsBatchFiles"

# Call the function with the correct argument
check_for_zero(folder_path)
```
